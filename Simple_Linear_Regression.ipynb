```python
import torch
# Mengimpor PyTorch untuk operasi tensor dan autograd

import torch.nn as nn
# Mengimpor modul neural network PyTorch (layer, loss, model)

X = torch.tensor([[1], [2], [3], [4]], dtype=torch.float32)
# Data input (fitur) berbentuk matriks 2D
# Setiap baris adalah satu sampel, satu fitur per sampel

Y = torch.tensor([[2], [4], [6], [8]], dtype=torch.float32)
# Target/output sesuai fungsi y = 2x

n_samples, n_features = X.shape
# Mengambil jumlah sampel dan jumlah fitur dari X

print(n_samples, n_features)
# Menampilkan jumlah data dan fitur

X_test = torch.tensor([5], dtype=torch.float32)
# Data uji untuk melakukan prediksi setelah training

input_size = n_features
# Jumlah neuron pada layer input

output_size = n_features
# Jumlah neuron pada layer output
# (karena model memetakan 1 input → 1 output)

# model = nn.Linear(input_size, output_size)
# Cara singkat membuat model linear (tanpa class)

class LinearRegression(nn.Module):
  # Mendefinisikan model custom dengan mewarisi nn.Module

  def __init__(self, input_dim, output_dim):
    super(LinearRegression, self).__init__()
    # Memanggil constructor parent class nn.Module

    self.lin = nn.Linear(input_dim, output_dim)
    # Mendefinisikan layer linear (y = wx + b)

  def forward(self, x):
    return self.lin(x)
    # Forward pass model

model = LinearRegression(input_size, output_size)
# Membuat instance model linear regression

print(f'Prediction before training: f(5) = {model(X_test).item():.3f}')
# Menampilkan prediksi sebelum training
# Nilai masih acak karena bobot belum dilatih

learning_rate = 0.01
# Learning rate untuk optimizer

n_iters = 100
# Jumlah epoch training

loss = nn.MSELoss()
# Mendefinisikan loss function Mean Squared Error

optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
# Mendefinisikan optimizer Stochastic Gradient Descent
# optimizer akan mengupdate parameter model (w dan b)

for epoch in range(n_iters):
    # Loop training

    y_pred = model(X)
    # Forward pass: prediksi model untuk seluruh data

    l = loss(Y, y_pred)
    # Menghitung loss

    l.backward()
    # Backpropagation
    # Menghitung gradien semua parameter model

    optimizer.step()
    # Update bobot dan bias menggunakan gradien

    optimizer.zero_grad()
    # Mengosongkan gradien agar tidak terakumulasi

    if epoch % 10 == 0:
      # Menampilkan hasil setiap 10 epoch

      [w, b] = model.parameters()
      # Mengambil bobot dan bias dari layer linear

      print(f'epoch {epoch+1} : w = {w[0][0].item():.3f}, loss = {l:.8f}')

print(f'Prediction after training: f(5) = {model(X_test).item():.3f}')
# Menampilkan prediksi setelah training
# Nilai mendekati 10 karena model belajar y ≈ 2x

# RINGKASAN:
# Kode ini mengimplementasikan linear regression lengkap menggunakan PyTorch:
# - Model dibuat dengan nn.Module dan nn.Linear
# - Loss function menggunakan Mean Squared Error
# - Optimisasi menggunakan SGD
# - Training loop mengikuti standar deep learning (forward → loss → backward → update)
# Ini adalah pola dasar semua training neural network di PyTorch.
```
