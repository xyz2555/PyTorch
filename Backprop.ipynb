```python
import torch
```

```python
x = torch.tensor(1.0)
# Membuat tensor skalar x dengan nilai 1.0
# Tidak membutuhkan gradien karena dianggap sebagai data input

y = torch.tensor(2.0)
# Membuat tensor skalar y dengan nilai 2.0
# Berperan sebagai target / ground truth

w = torch.tensor(1.0, requires_grad=True)
# Membuat parameter model w (weight) dengan nilai awal 1.0
# requires_grad=True agar PyTorch menghitung gradien terhadap w

y_hat = w * x
# Forward pass model sederhana: y_hat = w · x
# Ini adalah model linear paling dasar (1 parameter)

loss = y_hat - y
# Menghitung error (selisih antara prediksi dan target)

loss = loss ** 2
# Mengkuadratkan error → Mean Squared Error (tanpa mean karena 1 data)

print(loss)
# Menampilkan nilai loss (kesalahan model)

loss.backward()
# Melakukan backpropagation
# Menghitung ∂loss / ∂w dan menyimpannya di w.grad

print(w.grad)
# Menampilkan gradien weight w
# Nilai ini menunjukkan arah dan besar perubahan w untuk memperkecil loss

# RINGKASAN:
# Kode ini menunjukkan contoh paling dasar training model di PyTorch:
# model linear (y = w·x), perhitungan loss (squared error),
# serta penggunaan autograd untuk menghitung gradien parameter w
# sebagai dasar optimisasi dalam machine learning.
```
